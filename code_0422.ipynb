# Description

- This code is created for preprocessing data
---
    Use superstore's sales data as example,
    if each row of the data means one sales' record, and now we want to know how many sales in specific category(EX: Column "Catefory" in this data) each month,
    we can use this code to get it.


```python
df = pd.read_excel("./super_store_excel.xlsx")
display(df.head())
print(len(df))
```

```python
def countDateMonthly(df, col1, rename_col):
    """計算每月交易次數
    col1=           時間欄位                EX: order date
    reanme_col=     計算後的欄位名稱         EX: 訂單次數
    """
    df[col1] = pd.to_datetime(df[col1])
    df['月份'] = df[col1].dt.to_period('M') # 新增一個月份欄位
    orders_count = df.groupby('月份').agg({col1: 'count',}) # 按月份分組，計算訂單和出貨次數
    orders_count = orders_count.reset_index()
    orders_count['月份'] = orders_count['月份'].astype(str).str.slice(stop=7)
    # rename count column
    orders_monthly_count = orders_count.rename(columns={col1:rename_col})
    return orders_monthly_count

def countCategoryDateMonthly(df, col1, col2):
    """計算每月交易次數
    col1=           時間欄位                EX: order date
    col2=           類別欄位                EX: category
    """
    df[col1] = pd.to_datetime(df[col1])
    df['月份'] = df[col1].dt.to_period('M') # 新增一個月份欄位
    orders_count = df.pivot_table(index='月份', columns=col2, aggfunc='size', fill_value=0)
    orders_count.reset_index(inplace=True)
    return orders_count
    
```

```python
categortMonthly = countCategoryDateMonthly(df, "order_date", "Category")
categortMonthly
```
